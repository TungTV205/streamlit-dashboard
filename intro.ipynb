```python
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

plt.plot(x, y, label='tung', color='green')

# Tùy chỉnh đồ thị
plt.title('Sine Wave Example') # Tiêu đề
plt.xlabel('X-axis') # Nhãn trục X
plt.ylabel('Y-axis') # Nhãn trục Y
plt.legend() # Hiển thị chú thích của label 
plt.show()
```

```python
# Tạo biểu đồ phân tán

# Tao du lieu
np.random.seed(42) # Đặt seed để tái tạo kết quả
x = np.random.rand(50) # Tạo 50 giá trị ngẫu nhiên từ 0 đến 1

# tinh gia tri y theo phuong trinh y = 2x + 1 + ngẫu nhiên
y = 2 * x + 1 + np.random.randn(50) * 0.1

# ve biểu đồ phân tán
plt.scatter(x, y, label='Dữ liệu phân tán', color='green', marker='o', s=50, alpha=0.8) 
plt.show()
```
